```python
df_prep = pd.read_csv("game_news.csv", encoding='UTF8', sep=",")
df_prep.head(10) #/tail
```

```python
df_prep['text_lower']  = df_prep['text'].str.lower()
df_prep['text_lower'].head()
```

```python
# df_prep
```

```python
from nltk.corpus import stopwords
STOPWORDS = set(stopwords.words('russian'))
def stopwords(text):
    return " ".join([word for word in str(text).split() if word not in STOPWORDS])
# Применим функцию удаления стоп-слов
# df["text_stop"] = df["text_punct"].apply(stopwords)
print(STOPWORDS)
```

```python
df_prep["text_stop"] = df_prep["text_lower"].apply(stopwords)
df_prep["text_stop"].head()
```

```python
tokenizer = RegexpTokenizer(r'\w+')

df_prep['text_tok'] = df_prep['text_stop']
i = 0
for news in df_prep['text_tok']:
    df_prep['text_tok'][i] = tokenizer.tokenize(df_prep['text_tok'][i])
    i += 1
# df_prep["text_tok"] = df_prep['text_stop'].tokenizer.tokenize()
df_prep['text_tok']
```

```python
m = mystem()

df_prep['text_lem'] = df_prep['text_stop']
j = 0
for new in df_prep['text_lem']:
    df_prep['text_lem'][j] = m.lemmatize(new)
    j += 1
print(df_prep['text_lem'])
```

```python
print(df_prep['text_lem'])
```

```python
df_prep['text_lemmas'] = df_prep['text_lem']
k = 0
for text_new in df_prep['text_lemmas']:
    df_prep['text_lemmas'][k] = (''.join(text_new))
    k += 1
print(df_prep['text_lemmas'])
```

```python
df_prep['text_lemmas']
```

```python
df_prep['text_punct'] = df_prep['text_lemmas']

n = 0
for sent in df_prep['text_punct']:
    df_prep['text_punct'][n] = re.sub(r'[^\w\s\d+]','', sent)
    n += 1
    
df_prep['text_punct'].head()
```

```python
from gensim.models import Word2Vec
from gensim.models.word2vec import LineSentence
```

```python
w2v_model = Word2Vec(
    min_count=10,
    window=2,
    vector_size=300,
    negative=5,
    alpha=0.03,
    min_alpha=0.0007,
    sample=6e-5,
    sg=1)
#vector_size — размер векторного представления слова (word embedding).
#negative — сколько неконтекстных слов учитывать в обучении, используя negative sampling.
#alpha — начальный learning_rate, используемый в алгоритме обратного распространения ошибки (Backpropogation).
#min_alpha — минимальное значение learning_rate, на которое может опуститься в процессе обучения.
#sg — если 1, то используется реализация Skip-gram; если 0, то CBOW.
```

```python
#Получаем лист слов
from nltk.tokenize import sent_tokenize
from gensim.utils import simple_preprocess

class MySentences(object):
    def __init__(self, docs):
        self.corpus = docs
    def __iter__(self):
        for doc in self.corpus:
            doc_sentences = sent_tokenize(doc)
            for sent in doc_sentences:
                yield simple_preprocess(sent) # yields a tokenized 
```

```python
sentences = MySentences(df_prep['text_punct'].tolist())
```

```python
#Получаем словарь
w2v_model.build_vocab(sentences)
```

```python
# import nltk
# nltk.download('punkt_tab')
```

```python
w2v_model.train(sentences, total_examples=w2v_model.corpus_count, epochs=6, report_delay=1)
```

```python
#для сохранения оперативной памяти можно написать следующее
w2v_model.init_sims(replace=True)
```
